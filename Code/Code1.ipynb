```python
# Define the News API URL and API Key
api_url_News = 'https://newsapi.org/v2/everything'
api_key = 'cf94a3fb1f6246cd9fed77800990a6f7'

# Example query to get news on women entrepreneurs
params = {
    'q': 'women entrepreneurs',
    'apiKey': api_key
}

# Send GET request
response = requests.get(api_url_News, params=params)
data = response.json()

# Check if the response contains articles
if data['status'] == 'ok' and 'articles' in data:
    articles = data['articles']
    
    # Define CSV file name and headers
    csv_file = 'women_entrepreneurs_news.csv'
    headers = ['Source', 'Author', 'Title', 'Description', 'URL', 'Image URL', 'Published At', 'Content']
    
    # Open the CSV file in write mode
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=headers)
        
        # Write headers to the CSV file
        writer.writeheader()
        
        # Loop through each article and write relevant information to the CSV file
        for article in articles:
            # Extract relevant fields
            source = article.get('source', {}).get('name', 'N/A')  # 'name' is inside the 'source' dictionary
            author = article.get('author', 'N/A')
            title = article.get('title', 'N/A')
            description = article.get('description', 'N/A')
            url = article.get('url', 'N/A')
            image_url = article.get('urlToImage', 'N/A')
            published_at = article.get('publishedAt', 'N/A')
            content = article.get('content', 'N/A')
            
            # Write the data row to CSV
            writer.writerow({
                'Source': source,
                'Author': author,
                'Title': title,
                'Description': description,
                'URL': url,
                'Image URL': image_url,
                'Published At': published_at,
                'Content': content
            })
    
    print(f"Data has been written to {csv_file}")
else:
    print("No articles found or error with the API response")
```
